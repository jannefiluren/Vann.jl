# Snowfall and rainfall equation

```julia
using Gadfly;
```

```julia
# Parameters

tth_phase = 0.;
m_phase = 0.2;

# Air temperature

tair = collect(-10:0.01:10);

# Compute fraction snowfall

frac_snowfall = 1. ./ (1. + exp( (tair - tth_phase) ./ m_phase ));

plot(x = tair, y = frac_snowfall, Geom.line,
Guide.xlabel("Air temperature"), Guide.ylabel("Fraction snowfall"))
```

```julia
# Combine this into one function with two outputs

function split_prec(prec::Float64, tair::Float64, tth_phase::Float64, m_phase::Float64)
    
    frac_snowfall = 1. / (1. + exp( (tair - tth_phase) / m_phase ));
    
    psolid = prec * frac_snowfall;
    pliquid = prec - psolid;
    
    return psolid, pliquid
    
end

function test()
    for i in 1:(10*365*12)
        psolid, pliquid =  split_prec(10.0, -10.0, 0.0, 0.5)
    end
end

test()

@time test()
```

```julia
# Parameters

prec = 10.;
tth_phase = 0.;
m_phase = 0.2;

# Air temperature

tair = collect(-10:0.01:10);

# Compute solid and liquid precipitation

psolid = [];
pliquid = [];

for tair_in in tair
    psolid_tmp, pliquid_tmp = split_prec(prec, tair_in, tth_phase, m_phase);
    push!(psolid, psolid_tmp);
    push!(pliquid, pliquid_tmp);
end

plot(x = tair, y = psolid, Geom.line,
Guide.xlabel("Air temperature"), Guide.ylabel("Solid precipitation"))
```

```julia
plot(x = tair, y = pliquid, Geom.line,
Guide.xlabel("Air temperature"), Guide.ylabel("Liquid precipitation"))
```

# Snowmelt equation

Note that parameter m_melt should not be small (<0.2) otherwise numerical errors may aries

```julia
# Parameters

ddf = 4.;
tth_melt = 0.;
m_melt = 0.2;

# Air temperature

tair = collect(-20:0.01:20);

# Compute melt

t_m = (tair - tth_melt) ./ m_melt;
melt = ddf * m_melt .* (t_m + log(1. + exp(-t_m)));

plot(x = tair, y = melt, Geom.line, Guide.xlabel("Air temperature"), Guide.ylabel("Snowmelt"))
```

```julia
# Write as a function

function compute_melt(tair::Float64, ddf::Float64, tth_melt::Float64, m_melt::Float64)

    t_m = (tair - tth_melt) / m_melt;
    melt = ddf * m_melt * (t_m + log(1 + exp(-t_m)));

    return melt
    
end

function test()
    for i in 1:(10*365*12)
        melt =  compute_melt(10.0, 3., 0.0, 0.2);
    end
end

test()

@time test() 
```

```julia
# Parameters

ddf = 4.;
tth_melt = 0.;
m_melt = 0.2;

# Air temperature

tair = collect(-10:0.01:10);

# Compute melt

melt = [];

for tair_in in tair
    melt_tmp = compute_melt(tair_in, ddf, tth_melt, m_melt);
    push!(melt, melt_tmp);
end

plot(x = tair, y = melt, Geom.line,
Guide.xlabel("Air temperature"), Guide.ylabel("Snowmelt"))
```

```julia
# Function for computing ddf

function compute_ddf(date::Date, ddf_min::Float64, ddf_max::Float64)
    
    doy = Dates.dayofyear(date);
    
    ddf = (ddf_max - ddf_min) * (0.5*sin(2.0*pi*(Dates.dayofyear(date) - 80)/366) + 0.5) + ddf_min;
    
    return ddf   
    
end

# Test computation

ddf_min = 3.0;
ddf_max = 6.0;


date = Date(2015,1,1);

iday = collect(1:700);
ddf  = zeros(length(iday));

for i in iday
    
    ddf[i] = compute_ddf(date, ddf_min, ddf_max);
    
    date = date + Dates.Day(1);
    
end

plot(x = iday, y = ddf, Geom.line, Guide.xlabel("Day"), Guide.ylabel("Degree-day factor"))
```

# Forumulation of a complete snow model

```julia

function compute_ddf(date::Date, ddf_min::Float64, ddf_max::Float64)
    
    doy = Dates.dayofyear(date);
    
    ddf = (ddf_max - ddf_min) * (0.5*sin(2.0*pi*(Dates.dayofyear(date) - 80)/366) + 0.5) + ddf_min;
    
    return ddf   
    
end

function split_prec(prec::Float64, tair::Float64, tth_phase::Float64)
    
    m_phase = 1.0;
    
    frac_snowfall = 1. / (1. + exp( (tair - tth_phase) / m_phase ));
    
    psolid = prec * frac_snowfall;
    pliquid = prec - psolid;
    
    return psolid, pliquid
    
end


function compute_melt(tair::Float64, fw::Float64, ddf::Float64, tth_melt::Float64)
    
    m_melt = 0.5;

    t_m = (tair - tth_melt) / m_melt;
    melt = ddf * m_melt * (t_m + log(1. + exp(-t_m)));
    
    melt = min(max(0.,fw), melt);

    return melt
    
end


function compute_refreeze(tair::Float64, lw::Float64, ddf::Float64, tth_refr::Float64)

    m_refr = 0.5;

    t_m = (tth_refr - tair) / m_refr;
    refr = ddf * m_refr * (t_m + log(1. + exp(-t_m)));    
    
    if refr > lw
        refr = lw;
    end
    
    refr = min(lw,refr);
    
end
```

```julia
# Generate some dummy temperature data

date = collect(Date(2010):Date(2016));

tair = 7*sin(2.0*pi*(Dates.dayofyear(date) - 80)/366) + 3*randn(length(date))

plot(x = date, y = tair, Geom.line)
```

```julia
# Generate some dummy precipitation data

prec = rand(-40:0.01:10,length(date));

prec[prec .< 0] = 0

plot(x = date, y = prec, Geom.line)
```

```julia
# Parameters

tth = 0.;
ddf_min = 3.;
ddf_max = 6.;
whcap = 0.04;


# Initial states

swe_ini = 0.;
lw_ini = 0.;

swe = swe_ini;
lw = lw_ini;

# Outputs

swe_out = zeros(length(tair));
lw_out = zeros(length(tair));
infilt_out = zeros(length(tair));


# Loop over time

for i in eachindex(date)
    
    # Compute degree-day factor
    
    ddf = compute_ddf(date[i], ddf_min, ddf_max);
    
    # Compute solid and liquid precipitation

    psolid, pliquid = split_prec(prec[i], tair[i], tth);
    
    # Compute solid water in snow (mm)
    
    fw  = swe - lw;
    
    # Constrain fw and lw

    if fw < 0.
        fw = 0.;
    end
    
    if lw < 0.
        lw = 0.;
    end
    
    # Compute melt rates
    
    melt = compute_melt(tair[i], fw, ddf, tth);
    
    # Compute refreezing rates

    refr = compute_refreeze(tair[i], lw, ddf, tth);
    
    # Massbalance for frozen water in snowpack
    
    dfw = psolid + refr - melt;
    fw  = fw + dfw;
    
    # Massbalance for liquid water in snowpack
    
    dlw = pliquid + melt - refr;
    lw  = lw + dlw;

    lwmax = fw * whcap / (1 - whcap);
    
    infilt = lw - lwmax;

    if infilt < 0.
        infilt = 0.;
    end
    
    lw = lw - infilt;
    
    # Massbalance for snowpack
    
    swe = fw + lw;
    
    # Save outputs
    
    swe_out[i] = swe;
    lw_out[i] = lw;
    infilt_out[i] = infilt;
    
    
end


dswe = swe - swe_ini;
dlw = lw - lw_ini;

inputs_sum = sum(prec);
outputs_sum = sum(infilt_out);

massbalance = dswe + dlw - inputs_sum + outputs_sum;

println("Massbalance error = $massbalance mm")

```

```julia
set_default_plot_size(20cm, 15cm)

plot(layer(x = date, y = swe_out, Geom.line, Theme(default_color=colorant"orange")),
layer(x = date, y = lw_out, Geom.line), Scale.y_continuous(scalable=false))
```
